Q3. Explain the difference between instance methods and class methods. Provide an example of each?

Ans3. Instance Methods:


           Definition: Instance methods operate on instances of a class. They can access and modify the instance's attributes (data) and are called on an object of the class.
           
           Syntax: In Python, instance methods take self as their first parameter, which refers to the instance of the class.

           eg:

             class Dog:
               def __init__(self, name):
                 self.name = name

               def bark(self):
                 return f"{self.name} says woof!"

            # Creating an instance of Dog
            my_dog = Dog("Rex")
            print(my_dog.bark())  # Output: Rex says woof!


Class Methods:
     
          Definition: Class methods belong to the class itself rather than to any specific instance. They can access class-level data and are often used for factory methods or to maintain state related to the class as a whole.

          Syntax: In Python, class methods take cls as their first parameter and are decorated with @classmethod.

          eg:

            class Dog:
              species = "Canis lupus familiaris"  # Class attribute

              def __init__(self, name):
                self.name = name

              @classmethod
              def get_species(cls):
                return cls.species

             # Calling the class method
             print(Dog.get_species())  # Output: Canis lupus familiaris

Q5.  What are the three types of access modifiers in Python? How are they denoted?


Ans5. Public:

          Denotation: No prefix.
          Access: Accessible from anywhere.
          Example: self.attribute

  Protected:

         Denotation: Single underscore prefix (_).
         Access: Intended for internal use, accessible but discouraged from outside.
         Example: self._attribute

  Private:

        Denotation: Double underscore prefix (__).
       Access: Intended to be inaccessible from outside the class.
       Example: self.__attribute


Q6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance?

Ans6. Single Inheritance:

          A class (subclass) inherits from one parent class (superclass).

Multiple Inheritance:

         A class inherits from multiple parent classes.
         EG:

           class Canine:
             def bark(self):
               return "Bark!"

           class Pet:
             def play(self):
               return "Playing!"

           class Dog(Canine, Pet):
             pass

          dog = Dog()
          print(dog.bark())  # Output: Bark!
          print(dog.play())  # Output: Playing!

Multilevel Inheritance:

       A class inherits from a subclass, creating a chain of inheritance.
          

Hierarchical Inheritance:

      Multiple subclasses inherit from a single parent class.

Hybrid Inheritance:

      A combination of two or more types of inheritance.

Q8. Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses
`Circle` and `Rectangle` that implement the `area()` method.?

      import abc
      class Shape:
        @abc.abstractmethod
        def area(self):
          pass

      class Circle(Shape):
        def __init__(self, radius):
          self.radius = radius
        def area(self):
          return 3.14 * self.radius * self.radius
      class Rectangle(Shape):
        def __init__(self , l , b):
          self.length=l
          self.breath=b
        def area(self):   
          return self.length * self.breath

Q13.Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

Ans13.

Diamond Problem: Ambiguity arises in multiple inheritance when a class inherits from two classes that share a common ancestor.


Resolution: Python uses the C3 linearization algorithm to determine the Method Resolution Order (MRO), ensuring that the inheritance structure is respected and avoiding ambiguity in method calls. You can check the MRO using the mro() method or __mro__ attribute. For example, D.mro() will show the order of class resolution.
